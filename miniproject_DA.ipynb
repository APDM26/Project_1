```python
#Q2. Max, min , mean of asked amount,asked equity, asked valuation? 

max_pitcher_ask_amount = df['pitcher_ask_amount'].max()
min_pitcher_ask_amount = df['pitcher_ask_amount'].min()
mean_pitcher_ask_amount = df['pitcher_ask_amount'].mean()

# Calculate max, min, and mean of asked_equity
max_ask_equity = df['ask_equity'].max()
min_ask_equity = df['ask_equity'].min()
mean_ask_equity = df['ask_equity'].mean()

# Calculate max, min, and mean of asked_valuation
max_ask_valuation = df['ask_valuation'].max()
min_ask_valuation = df['ask_valuation'].min()
mean_ask_valuation = df['ask_valuation'].mean()

# Print the results
print('Max pitcher_ask_amount:', max_pitcher_ask_amount)
print('Min pitcher_ask_amount:', min_pitcher_ask_amount)
print('Mean pitcher_ask_amount:', mean_pitcher_ask_amount)
print('Max ask equity:', max_ask_equity)
print('Min ask equity:', min_ask_equity)
print('Mean ask equity:', mean_ask_equity)
print('Max ask valuation:', max_ask_valuation)
print('Min ask valuation:', min_ask_valuation)
print('Mean ask valuation:', mean_ask_valuation)
```

```python
#Q3. Max & Min asked- equity,asked-valuation and asked amount episode-wise? 


# Group the data by episode



result1=df.groupby('episode_number')['ask_equity'].agg(['min', 'max'])
result2=df.groupby('episode_number')['ask_valuation'].agg(['min', 'max'])
result3=df.groupby('episode_number')['ask_valuation'].agg(['min', 'max'])


# Display the result
print("Ask_Equity_by_ep_no",result1)
print("ask_valuation_by_ep_no",result2)
print("ask_valuation_by_ep_no",result3)
```

```python
#Q4. Brand names in which 2 ,3 or 4 sharks are invested? 

filtered_df = df.loc[(df['total_sharks_invested']>=2)& (df['total_sharks_invested']<=4)]
sorted_df = filtered_df.sort_values(by='total_sharks_invested', ascending=False)
def left_align(x):
    return f"{x:<30}"

print(sorted_df[['brand_name','total_sharks_invested']].to_string(index=False,formatters={'brand_name': left_align}))


print(df.style.format(text_align='left_align'))
```

```python
#Q5. Episode wise minimum and maximum no. of sharks invested? 

result= df.groupby('episode_number')['total_sharks_invested'].agg(['min', 'max'])

# Display the result
print(result)
```

```python
#Q6. Find the brands in which only ashneer invested

filtered_df2 = df.loc[(df['ashneer_present']==1)& (df['total_sharks_invested']==1)]

def left_align(x):
    return f"{x:<30}"
print(filtered_df2[['brand_name','ashneer_present','deal_valuation']].to_string(index=False,formatters={'brand_name': left_align}))
```

```python
#Q7. Brand names who ask for 1 crore and got a deal what was the deal amount?

filtered_df5 = df.loc[df['pitcher_ask_amount']==100]
sorted_df3 = filtered_df5.sort_values(by='deal_amount', ascending=False)
def left_align(x):
    return f"{x:<30}"
print(sorted_df3[['brand_name','pitcher_ask_amount','deal_amount']].to_string(index=False,formatters={'brand_name': left_align}))
```

```python
#Q8. Brand names where deal equity is between 1 to 50?

filtered_df3 = df.loc[(df['deal_equity']>=1)&(df['deal_equity']<=50)]
sorted_df2 = filtered_df3.sort_values(by='deal_equity', ascending=False)
def left_align(x):
    return f"{x:<30}"




print(sorted_df2[['brand_name','deal_equity']].to_string(index=False,formatters={'brand_name': left_align}))
```
